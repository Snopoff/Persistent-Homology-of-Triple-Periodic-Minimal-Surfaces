```python
import numpy as np
import pyvista as pv
import csv

# количество генерируемых точек для поверхности. В дальнейшем задаваемый параметр
count_pount = 10
def generate_points(count_pount):
    one_axis = np.linspace(-np.pi, np.pi, count_pount)
    xyz_mesh = np.meshgrid(one_axis, one_axis, one_axis)
    return x,y,z
    

# генерация случайных чисел для парметра d
size = 20
np.random.seed(1999)
d_params = sorted(np.random.uniform(low=-0.99, high=0.99, size=size)) #В этом массиве все параметры d
print(d_params)
d_param = d_params[19] #ТУТ МЕНЯТЬ
print(d_param)

#временные названия файлов для сохранения
filename_for_Schwarz_p = 'pointcloud_Schwarz_p_19.csv' #ТУТ МЕНЯТЬ
filename_for_Gyroid = 'pointcloud_Gyroid_19.csv' #ТУТ МЕНЯТЬ


def generate_tpms_Schwarz_p(d_params):
    x,y,z = generate_points(count_pount)
    
    porosity_list = []
    pointcloud_list = []
    
    for i in range(len(d_params)):
        Schwarz_p = np.cos(x) + np.cos(y) + np.cos(z) + d_param #эт формула поверхности Шварца
        tolerance = 0.001
        x0,y0,z0 = np.where(abs(Schwarz_p)<tolerance)
        pointcloud = np.vstack((x0,y0,z0)).T
        cube_volume = (np.min(x0) + np.max(x0)) * (np.min(y0) + np.max(y0)) * (np.min(z0) + np.max(z0)) #Объем кубика
        cloud = pv.PolyData(pointcloud)
        surface = cloud.reconstruct_surface() # Облако точек для расчета объема
        surface_volume = surface.volume # Объем поверхности
        porosity = (cube_volume-surface_volume)/cube_volume #расчитываем удельную пористость
        porosity_list.append(porosity)
        pointcloud_list.append(pointcloud)
        

    with open(filename_for_Schwarz_p, mode='w', newline='') as file:
        writer = csv.writer(file)
        writer.writerows(pointcloud)

    return porosity_list, pointcloud_list

porosity_Schwarz_p = generate_tpms_Schwarz_p(d_param)
print(porosity_Schwarz_p)
porosity_list_Schwarz_p.append(porosity_Schwarz_p)


def generate_tpms_Gyroid(d_param):
    x,y,z = generate_points(count_pount)
    
    #porosity_list = []
    #pointcloud_list = []
    #for i in range(len(d_param)):
    Gyroid = np.cos(x)*np.sin(y) + np.cos(y)*np.sin(z) + np.cos(z)*np.sin(x) + d_param
    tolerance = 0.001
    x0, y0, z0 = np.where(abs(Gyroid) < tolerance)
    del Gyroid
    pointcloud = np.vstack((x0,y0,z0)).T
    cube_volume = (np.min(x0) + np.max(x0)) * (np.min(y0) + np.max(y0)) * (np.min(z0) + np.max(z0)) #Объем кубика
    del x0
    del y0
    del z0
    cloud = pv.PolyData(pointcloud)
    surface = cloud.reconstruct_surface() # Облако точек для расчета объема
    del cloud
    surface_volume = surface.volume # Объем поверхности
    del surface
    porosity = (cube_volume-surface_volume)/cube_volume #расчитываем удельную пористость
    #porosity_list.append(porosity)
    #pointcloud_list.append(pointcloud)
    with open(filename_for_Gyroid, mode='w', newline='') as file:
            writer = csv.writer(file)
            writer.writerows(pointcloud)

    return porosity

porosity_Gyroid = generate_tpms_Gyroid(d_param)

print(porosity_Gyroid)
porosity_list_Gyroid.append(porosity_Gyroid)
```
